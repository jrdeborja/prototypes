### **Correlation method**
- removal of highly correlated descriptors and fingerprint keys

```python
# mordred descriptors
corr_matrix = hcv_md.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Threshold > 0.80
to_drop = [column for column in upper.columns if any(upper[column] > 0.80)]

hcv_md = hcv_md.drop(to_drop, axis=1)
```

```python
hcv_md
```

```python
# maccs keys
corr_matrix = hcv_fp.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Threshold > 0.80
to_drop = [column for column in upper.columns if any(upper[column] > 0.80)]

hcv_fp = hcv_fp.drop(to_drop, axis=1)
```

```python
hcv_fp
```

```python
x_hcv = pd.concat([hcv_md, hcv_fp], axis=1)
y_hcv = hcv_pIC50
```

```python
x_hcv
```

```python
# final correlation run on x_hcv
corr_matrix = x_hcv.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Threshold > 0.80
to_drop = [column for column in upper.columns if any(upper[column] > 0.80)]

x_hcv = x_hcv.drop(to_drop, axis=1)
```

### **Hyperparameter tuning**
- using RandomizedSearchCV and GridSearchCV on NuSVR
- all of the samples from hcv dataset

```python
import numpy as np
from scipy.stats import uniform, randint
from joblib import parallel_backend

from sklearn.metrics import r2_score, mean_squared_error, make_scorer
from sklearn.svm import SVR
from sklearn.model_selection import RandomizedSearchCV

# Best parameter selection using RandomizedSearchCV
with parallel_backend('multiprocessing', n_jobs=-1):
  base_svr = SVR()

  # Parameters to be tuned for SVR
  param_distrib_svr = {
      'C' : uniform(1, 3),
      'tol' : uniform(0.001, 0.002),
      'epsilon' : uniform(0.1, 0.35)
  }

  # Selection of the best parameters for SVR
  rscv_svr = RandomizedSearchCV(base_svr, param_distributions=param_distrib_svr,
                                      n_jobs=-1, cv=5, random_state=42).fit(x_train, y_train)
```

```python
rscv_svr.best_params_
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C' : [rscv_svr.best_params_['C'] + 1,
           rscv_svr.best_params_['C'],
           rscv_svr.best_params_['C'] - 1],
    'epsilon' : [rscv_svr.best_params_['epsilon'] + 0.1,
                 rscv_svr.best_params_['epsilon'],
                 rscv_svr.best_params_['epsilon'] - 0.1],
    'tol' : [rscv_svr.best_params_['tol'] + 0.001,
             rscv_svr.best_params_['tol'],
             rscv_svr.best_params_['tol'] - 0.001]
}

base2_svr = SVR()

# Best parameter selection using GridSearchCV
with parallel_backend('multiprocessing', n_jobs=-1):
  gscv_svr = GridSearchCV(base2_svr, param_grid=param_grid,
                                n_jobs=-1, cv=5).fit(x_train, y_train)
```
